```python
import pandas as pd
```

```python
# Load the CSV file
df = pd.read_csv(r'C:\Users\Asus\superstore_sales.csv')
```

```python
df.head()
```

```python
# Convert order_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y')

# Create Month-Year column
df['Month-Year'] = df['order_date'].dt.to_period('M').astype(str)

# Optional: Sort by Order Date
df = df.sort_values('order_date')
```

```python
monthly_sales = df.groupby('Month-Year')['sales'].sum().reset_index()
```

```python
region_sales = df.groupby('region')['sales'].sum().reset_index()
```

```python
category_sales = df.groupby('category')['sales'].sum().reset_index()
```

```python
df.to_csv('cleaned_superstore_sales.csv', index=False)
monthly_sales.to_csv('monthly_sales.csv', index=False)
region_sales.to_csv('region_sales.csv', index=False)
category_sales.to_csv('category_sales.csv', index=False)
```
